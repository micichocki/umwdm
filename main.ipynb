# ChestMnist Dataset

```python
import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt
from PIL import Image
from medmnist import ChestMNIST

# use download=True Or download manually https://zenodo.org/records/10519652/files/chestmnist_224.npz?download=1

train_ds = ChestMNIST(root='./data', split='train', size=224)
test_ds = ChestMNIST(root='./data', split='test', size=224)
validation_ds = ChestMNIST(root='./data', split='val', size=224)
```

```python
print("Dataset length:", len(train_ds))

for i in range(5):
    img, label = train_ds[i]
    plt.imshow(img, cmap='gray')
    plt.grid(False)
    plt.axis('off')
    plt.show()
```

```python
def preprocess_image(image):
    processed_image = np.array(image)
    processed_image = processed_image / 255.0
    return processed_image
```

```python
sample_image, sample_label = train_ds[0]
plt.imshow(preprocess_image(sample_image), cmap='gray')
plt.grid(False)
plt.axis('off')
plt.show()
```
